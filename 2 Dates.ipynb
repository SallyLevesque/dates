## Create features for year, month, day, hour, and minute

```python
# Create features for year, month, day, hour, and minute

dates['year'] = dates['date'].dt.year
dates['month'] = dates['date'].dt.month
dates['day'] = dates['date'].dt.day
dates['dow_number'] = dates['date'].dt.dayofweek
dates['dow_name'] = dates['date'].dt.weekday_name
dates['is_month_end'] = dates['date'].dt.is_month_end


#remove extra date column if it exists
if 'date' in dates.columns:
    dates=dates.drop('date',axis=1)

# Show three rows
dates.head(3)
```
